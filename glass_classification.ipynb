```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

Fonte de Dados: https://www.kaggle.com/datasets/uciml/glass

```python
dados_glass = pd.read_csv('/content/drive/MyDrive/Teste_de_Python/glass_classification/glass.csv')
dados_glass
```

```python
X = dados_glass.drop(columns= ['Type'])
y = dados_glass['Type']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

# Inicialize o modelo
rf_model = RandomForestClassifier(random_state=42)

# Treine o modelo
rf_model.fit(X_train, y_train)

# Faça previsões
rf_predictions = rf_model.predict(X_test)

# Avalie o modelo
print("Random Forest Accuracy:", accuracy_score(y_test, rf_predictions))
print(classification_report(y_test, rf_predictions))
```

KNN

```python
from sklearn.neighbors import KNeighborsClassifier

#Inicializar o modelo
knn_model = KNeighborsClassifier(n_neighbors= 5)

# Treine o modelo
knn_model.fit(X_train, y_train)

# Faça previsões
knn_predictions = rf_model.predict(X_test)

# Avalie o modelo
print("K-Nearest Neighbors Accuracy:", accuracy_score(y_test, knn_predictions))
print(classification_report(y_test, knn_predictions))
```

Regressão Logistica

```python
from sklearn.linear_model import LogisticRegression

# Inicialize o modelo
log_model = LogisticRegression(max_iter=1000, multi_class='multinomial', solver='lbfgs')

# Treine o modelo
log_model.fit(X_train, y_train)

# Faça previsões
log_predictions = log_model.predict(X_test)

# Avalie o modelo
print("Logistic Regression Accuracy:", accuracy_score(y_test, log_predictions))
print(classification_report(y_test, log_predictions))
```

Rede Neural (MLP)

```python
from sklearn.neural_network import MLPClassifier

# Inicialize o modelo
mlp_model = MLPClassifier(hidden_layer_sizes=(100, 50), max_iter=1000, random_state=42)

# Treine o modelo
mlp_model.fit(X_train, y_train)

# Faça previsões
mlp_predictions = mlp_model.predict(X_test)

# Avalie o modelo
print("Neural Network (MLP) Accuracy:", accuracy_score(y_test, mlp_predictions))
print(classification_report(y_test, mlp_predictions))
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report

# Criando o modelo SVM
svm_model = SVC(random_state=42)

# Treinando o modelo
svm_model.fit(X_train, y_train)

# Prevendo as classes
y_pred_svm = svm_model.predict(X_test)

# Avaliando o desempenho
print("SVM Accuracy:", svm_model.score(X_test, y_test))
print(classification_report(y_test, y_pred_svm))
```

Árvore de Decisão

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report

# Criando o modelo Árvore de Decisão
tree_model = DecisionTreeClassifier(random_state=42)

# Treinando o modelo
tree_model.fit(X_train, y_train)

# Prevendo as classes
y_pred_tree = tree_model.predict(X_test)

# Avaliando o desempenho
print("Decision Tree Accuracy:", tree_model.score(X_test, y_test))
print(classification_report(y_test, y_pred_tree))
```
